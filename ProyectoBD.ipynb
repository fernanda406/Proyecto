```python
# Instalar libreras necesarias (si no estn ya instaladas)
!pip install matplotlib pandas

import pandas as pd
import matplotlib.pyplot as plt

# ==========================
# 1. Cargar datos desde CSV
# ==========================
clientes = pd.read_csv("clientes.csv")
productos = pd.read_csv("productos.csv")
pedidos = pd.read_csv("pedidos.csv")

# ==========================
# 2. Grficas
# ==========================

# Grfico 1: Distribucin de tipos de clientes
clientes['Tipo_Cliente'].value_counts().plot(kind='bar', color=['blue','orange','green'])
plt.title('Distribucin de Tipos de Clientes')
plt.xlabel('Tipo de Cliente')
plt.ylabel('Cantidad')
plt.show()

# Grfico 2: Precios de productos
productos.plot(x='Nombre', y='Precio', kind='bar', color='purple')
plt.title('Precios de Productos')
plt.xlabel('Producto')
plt.ylabel('Precio ($)')
plt.show()

# Grfico 3: Totales de pedidos
pedidos.groupby('Tipo_Pedido')['Total'].sum().plot(kind='bar', color='red')
plt.title('Totales de Pedidos por Tipo')
plt.xlabel('Tipo de Pedido')
plt.ylabel('Total ($)')
plt.show()
```
